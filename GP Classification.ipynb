```python
def numerically_stable_cdf(i):
    cdf_value = norm.cdf(i)
    if cdf_value == 0:
        return 1e-16
    elif cdf_value == 1:
        return 1-1e-16
    else:
        return cdf_value
```

Function we want to optimize:
$$\psi(\boldsymbol f) = \log p(\boldsymbol y | \boldsymbol f) - \frac{1}{2}\boldsymbol{f}^TK^{-1}\boldsymbol{f}-\frac{1}{2}\log|K| - \frac{n}{2}\log 2\pi$$

This depends on our choise for the likelihood, $p(\boldsymbol y | \boldsymbol f)$, which gives the target class probability as a function of the latent variable $f$. Say we use $p(y_i|f_i) = \Phi(y_if_i)$. Then

$$\nabla \psi(\boldsymbol f) = \nabla \log p(\boldsymbol y| \boldsymbol f) - K^{-1}\boldsymbol f$$
Where
$$\frac{y_i\mathcal{N}(f_i)}{\Phi(y_if_i)} - K^{-1}\boldsymbol f$$

And we have second derivative 
$$\nabla \nabla \psi(\boldsymbol f) = \nabla \nabla \log p(\boldsymbol y|\boldsymbol f)-K^{-1}$$
where
$$\frac{\partial^2}{\partial f^2_i} \log p(y_i|f_i) = -\frac{\mathcal{N}(f_i)^2}{\Phi(y_if_i)^2} - \frac{y_if_i\mathcal{N}(f_i)}{\Phi(y_if_i)}$$
So, at the maximum of $\psi(\boldsymbol f)$, we have
$$\nabla \psi = 0\Rightarrow \hat{\boldsymbol f} = K(\nabla \log p(\boldsymbol y|\hat{\boldsymbol f})$$

```python
def create_K(x,y,l,sig):
    n = len(x)
    K = np.zeros((n,n))
    for i in xrange(n):
        for j in xrange(n):
            K[i,j] = kernel(x,y,x[i],x[j],l,sig)
    return K
```

```python
def psi(x,y,f,K,K_inv):
    log_likelihood = loglik(x,y,f)
    return log_likelihood -.5*np.dot(np.dot(f.T,K_inv),f) #-.5*np.log(np.linalg.det(K))
```

```python
def loglik(x,y,f):
    probit_map = np.vectorize(numerically_stable_cdf)
    return np.sum(np.log(probit_map(np.multiply(f,y))))
```

```python
def kernel(x,y,i,j, l = 1, sig = 1):
    if type(i) == int:
        return sig*np.exp((-(i-j)**2)/l)
    else:
        return sig*np.exp((-np.linalg.norm(i-j)**2)/l)
```

```python
def loglik_deriv_i(x,y,f_i,i):
    return y[i]*norm.pdf(f_i)/(numerically_stable_cdf(y[i]*f_i))
```

```python
def loglik_deriv(x,y,f):
    n = len(x)
    deriv = np.array([loglik_deriv_i(x,y,f[i],i) for i in range(n)])
    return deriv
```

```python
def psi_deriv(x,y,f,K_inv):
    deriv = loglik_deriv(x,y,f)
    return deriv - np.dot(K_inv,f)
```

```python
def loglik_deriv_2_i(x,y,f_i,i):
    pdf_div_cdf = norm.pdf(f_i)/(numerically_stable_cdf(y[i]*f_i))
    return -pdf_div_cdf ** 2 - y[i]*f_i*pdf_div_cdf
```

```python
def create_W(x,y,f):
    n = len(x)
    W = np.zeros((n,n))
    for i in range(n):
        W[i,i] = loglik_deriv_2_i(x,y,f[i],i)
    return -1*W
```

```python
def optimize_f(x,y,K,K_inv):
    n = len(x)
    fnew = np.zeros(n)
    for epoch in range(100):
        f = fnew
        W = create_W(x,y,f)
        fnew = np.dot(np.linalg.inv(K_inv + W),np.dot(W,f)+loglik_deriv(x,y,f))
        if np.linalg.norm(f-fnew) < .00001:
            #print psi(x,y,f,K,K_inv)
            #print fnew
            break
        #print psi(x,y,f,K,K_inv)
    return fnew
```

```python
def create_K_star(x,y,x_new,l,sig):
    n = len(x)
    K_star = np.zeros((1,n))
    for i in xrange(n):
        K_star[0,i] = kernel(x,y,x_new,x[i],l,sig)
    return K_star
```

```python
def predict_and_error(x,y,x_new,f_hat,K,l,sig):
    K_star = create_K_star(x,y,x_new,l,sig)
    K_star_star = kernel(x,y,x_new,x_new,l,sig)
    W_hat = create_W(x,y,f_hat)
    hess_inv = np.linalg.inv(K + np.linalg.inv(W_hat))
    #pred_prob = numerically_stable_cdf(np.dot(K_star,loglik_deriv(x,y,f_hat))) MAP estimate
    pred_var = K_star_star - np.dot(np.dot(K_star,hess_inv),K_star.T)
    E_f_star = np.dot(K_star,loglik_deriv(x,y,f_hat))
    pi_star = numerically_stable_cdf(E_f_star/(np.sqrt(1+pred_var)))
    if pi_star.shape == (1,1):
        pi_star = pi_star[0][0]
    return(pi_star,pred_var)
```

```python
def plot_preds(x,y,l = 1,sig = 1):
    K = create_K(x,y,l,sig)
    K_inv = np.linalg.inv(K)
    f_hat = optimize_f(x,y,K,K_inv)
    xmin = np.min(x)-1
    xmax = np.max(x)+1
    p_x = np.arange(xmin,xmax,.1)
    p_y = (np.array([predict_and_error(x,y,x_new,f_hat,K,l,sig)[0] for x_new in p_x]))
    plt.plot(x,(y+1)/2,'bo',label='Sample observations')
    plt.plot(p_x,p_y,'r-',label='Regression line')
    plt.axhline(y=.5)
```

## Cholesky Optimization

```python
def optimize_f_cholesky(x,y,K,step_size = 1):
    n = len(x)
    fnew = np.zeros(n)
    for epoch in range(100):
        f = fnew
        W = create_W(x,y,f)
        W_sqrt = np.sqrt(np.abs(W)) # Absolute values to work around 0 -- values should never be negative
        L = np.linalg.cholesky(np.eye(n)+np.dot(np.dot(W_sqrt,K),W_sqrt))
        b = np.dot(W,f)+ step_size*loglik_deriv(x,y,f)
        a = b - np.dot(W_sqrt,np.dot(np.linalg.inv(L.T),np.dot(np.linalg.inv(L),np.dot(W_sqrt,np.dot(K,b)))))
        fnew = np.dot(K,a)
        if np.linalg.norm(f-fnew) < .00001:
            #print fnew
            break
        #print -.5*np.dot(a,fnew)+loglik(x,y,f)
    return fnew,W_sqrt,L
```

```python
def predict_and_error_cholesky(x,y,x_new,f_hat,W_sqrt,L,K,l,sig):
    n = len(x)
    K_star = create_K_star(x,y,x_new,l,sig)
    K_star_star = kernel(x,y,x_new,x_new,l,sig)
    #W_hat = create_W(x,y,f_hat)
    #W_sqrt = np.sqrt(np.abs(W_hat)) # Absolute values to work around 0 -- values should never be negative
    #L = np.linalg.cholesky(np.eye(n)+np.dot(np.dot(W_sqrt,K),(W_sqrt)))
    f = np.dot(K_star,loglik_deriv(x,y,f_hat))
    v = np.dot(np.linalg.inv(L),np.dot(W_sqrt,K_star.T))
    V = K_star_star-np.dot(v.T,v)
    pi_star = numerically_stable_cdf(f/(np.sqrt(1+V)))
    if pi_star.shape == (1,1):
        return pi_star[0][0]
    return pi_star
```

```python
def plot_preds_cholesky(x,y,l = 1,sig = 1,step_size = 1):
    K = create_K(x,y,l,sig)
    f_hat,W_sqrt,L = optimize_f_cholesky(x,y,K,step_size)
    xmin = np.min(x)-1
    xmax = np.max(x)+1
    p_x = np.arange(xmin,xmax,.1)
    p_y = (np.array([predict_and_error_cholesky(x,y,x_new,f_hat,np.abs(W_sqrt),L,K,l,sig) for x_new in p_x]))
    plt.plot(x,(y+1)/2,'bo',label='Sample observations')
    plt.plot(p_x,p_y,'r-',label='Regression line')
    plt.axhline(y=.5)
```

## 2-dimensional example

Try changing values of l. As l decreases, the decision boundaries become less conservative and perhaps more prone to overfitting.

```python
def plot_preds_cholesky_2d(X,y,l=1,sig=1,step_size=1):
    K = create_K(X,y,l,sig)
    f_hat,W_sqrt,L = optimize_f_cholesky(X,y,K,step_size)
    xmin = np.min(X[:,0])
    xmax = np.max(X[:,0])
    ymin = np.min(X[:,1])
    ymax = np.max(X[:,1])
    p_x = np.arange(xmin-.1,xmax+.1,.03)
    p_y = np.arange(ymin-.1,ymax+.1,.03)
    p_x,p_y = np.meshgrid(p_x,p_y)
    C_classes = np.zeros(p_x.shape)
    for i in xrange(len(p_x)):
        for j in xrange(len(p_x[0])):
            xi = p_x[i][j]
            xj = p_y[i][j]
            #C_classes[i][j] = predict_and_error_cholesky(X,y,np.array([xi,xj]),f_hat,np.abs(W_sqrt),L,K,l,sig)
            C_classes[i][j] = round(predict_and_error_cholesky(X,y,np.array([xi,xj]),f_hat,np.abs(W_sqrt),L,K,l,sig),-0) # for decision boundary
    plt.contourf(p_x, p_y, C_classes, cmap=plt.cm.bone)
    plt.scatter(*zip(*X[np.where(y == -1)[0]]),linewidth=1,s=50,c='red')
    plt.scatter(*zip(*X[np.where(y == 1)[0]]),linewidth=1,s=50,c='yellow')
```

## MNIST

```python
import sys,struct
import gzip
from array import array
```

```python
f = gzip.open('./mnist/t10k-images-idx3-ubyte.gz', 'rb') # test set
_, n_images, rows, cols = struct.unpack('>IIII', f.read(16)) # first 16 bytes are metadata
pixels = np.array(array('B', f.read())) / 255.0 # pixel data are unsigned bytes; normalize to [0,1] scale
f.close()
images = pixels.reshape((n_images, rows * cols)) # each image is a vector of 28*28 values in [0,1]
```

```python
digits=np.arange(10)
flbl = gzip.open('./mnist/t10k-labels-idx1-ubyte.gz', 'rb')
magic_nr, size = struct.unpack(">II", flbl.read(8))
lbl = array("b", flbl.read())
flbl.close()

fimg = gzip.open('./mnist/t10k-images-idx3-ubyte.gz', 'rb')
magic_nr, size, rows, cols = struct.unpack(">IIII", fimg.read(16))
img = array("B", fimg.read())
fimg.close()

ind = [ k for k in range(size) if lbl[k] in digits ]
N = len(ind)

labels = np.zeros((N, 1))
for i in range(len(ind)):
    labels[i] = lbl[ind[i]]
```

```python
def show_image(im):
    plt.imshow(1-im.reshape((rows,cols)),cmap=plt.cm.gray)
```

```python
def make_mnist_data(n_class,class_0,class_1):
    class_0s = images[np.where(labels == class_0)[0]]
    class_0s_train = class_0s[0:n_class]
    class_1s = images[np.where(labels == class_1)[0]]
    class_1s_train = class_1s[0:n_class]
    X = np.vstack((class_0s_train,class_1s_train))
    y = np.concatenate((-1*np.ones(n_class),np.ones(n_class)))
    class_0s_test = class_0s[n_class:,]
    class_1s_test = class_1s[n_class:,]
    X_test = np.vstack((class_0s_test,class_1s_test))
    y_test = np.concatenate((-1*np.ones(len(class_0s_test)),np.ones(len(class_1s_test))))
    return X,y,X_test,y_test
```

```python
def predict_mnist(X,y,X_test,l,sig):
    preds = np.zeros(len(X_test))
    K = create_K(X,y,l,sig)
    f_hat,W_sqrt,L = optimize_f_cholesky(X,y,K,1)
    for ind in xrange(len(X_test)):
        preds[ind] = 2*round(predict_and_error_cholesky(X,y,X_test[ind],f_hat,W_sqrt,L,K,l,sig),-0)-1
    return preds
```

```python
def get_predictions_and_test(n_class,class_0,class_1,l = np.exp(2.85)**2*2,sig = np.exp(2.35)**2):
    X,y,X_test,y_test = make_mnist_data(n_class,class_0,class_1)
    preds = predict_mnist(X,y,X_test,l,sig)
    return preds,y_test
```

```python
def get_accuracy(preds,y_test):
    return 1-np.mean(np.abs(preds-y_test)/2)
```

```python
def get_mistakes(preds,y_test):
    return np.where(np.abs(preds-y_test) == 2)[0]
```

```python
def plot_mistakes(preds,y_test):
    mistakes = get_mistakes(preds,y_test)
    random_mistakes = np.random.shuffle(mistakes)
    plt.figure(figsize=(10, 10))
    plt.suptitle('Mistakes', fontsize=24)
    for k in xrange(16):
        plt.subplot2grid((4, 4), (k / 4, k % 4))
        plt.axis('off')
        plt.imshow(1.0 - X_test[mistakes][k].reshape((rows, cols)), cmap=plt.cm.gray)
    plt.show()
```

```python
sizes = np.array([5,10,15,20,30,40,50,60,70,80,90,100])
accuracy = [get_accuracy(*get_predictions_and_test(size,1,8)) for size in sizes]
plt.title('MNIST Classification, Digits 1/8', fontsize=20)
plt.scatter(sizes,accuracy)
plt.xlabel("Size of train set")
plt.ylabel("Test accuracy")
```

```python
X,y,X_test,y_test = make_mnist_data(100,1,8)
preds = predict_mnist(X,y,X_test,l,sig)
plot_mistakes(preds,y_test)
```
